```python
import numpy as np
import pandas as pd
import os
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import random
%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
# import file

dir=r"E:\Kaggle_data\Loan_predict"
train=pd.read_csv(os.path.join(dir,"train.csv"))
test=pd.read_csv(os.path.join(dir,"test.csv"))
sub_=pd.read_csv(os.path.join(dir,"sample_submission.csv"))
orig=pd.read_csv(os.path.join(dir,"loan_dataset_20000.csv"))
```

```python
# revise orig data
orig["id"]=test.id.max()+1
test["loan_paid_back"]=0

display(f"shape of train<{train.shape}>")
display(f"shape of test<{test.shape}>")
display(f"shape of orig<{orig.shape}>")

data=pd.concat([train,test],axis=0)
display(f"shape of merged data <{data.shape}>")
```

```python
Features=train.columns
Features
```

```python
# display data

CAT,NUM,TARGET=[],[],["loan_paid_back"]
for col in data.columns.drop(["id","loan_paid_back"]):
    temp="NUM"
    if data[col].dtype=="object":
        CAT.append(col)
        temp="CAT"
    else:
        NUM.append(col)
    print(f"{col:20}[{temp}] {data[col].nunique():10} catgory{data[col].isna().sum():10} NAN")
Features=train.columns.drop(["id","loan_paid_back"]).tolist()

len(CAT),len(NUM),len(Features)
```

```python
# establish baseline with lightgbm

from lightgbm import LGBMClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score

lgb_params={
    "colsample_bytree":0.8,
    "learning_rate":0.06,
    "objective":"binary",
    "metric":"auc",
    "max_depth":8,
    "min_child_samples":59,
    "is_unbalance":True,
    "n_estimators":2000,
    "subsample":0.76,
    "verbose":500,
    "random_state":42
}
```

```python
def set_model(model_name,params):
    match model_name:
        case "lgb":
            from lightgbm import LGBMClassifier
            return LGBMClassifier(**params)
        case _:
            raise ValueError(f"Unknown Name < {mdoel_name} >")
```

```python
def cv_train(model_name,params,enable_catrgorical=True,SEED_5=5,SEED=42,cv=7,train=train.copy(),test=test.copy()):
    oof_preds=np.zeros(len(train))
    test_preds=np.zeros(len(test))
    random.seed(SEED)
    seeds=random.sample(np.arange(len(train)).tolist(),SEED_5)

    kf=StratifiedKFold(n_splits=cv,random_state=SEED,shuffle=True)
    for i,(train_idx,val_idx) in enumerate(kf.split(train,train[TARGET])):
        print(f"< {i+1}/{cv} > is training !")
        if enable_catrgorical:
            train[CAT]=train[CAT].astype("category")
            test[CAT]=test[CAT].astype("category")
        
        x_train,y_train=train.iloc[train_idx][Features],train.iloc[train_idx][TARGET]
        x_val,y_val=train.iloc[val_idx][Features],train.iloc[val_idx][TARGET]
        x_test=test[Features]
        
        # 5 seed trainings â€”â€” Ensure stable results
        for _ in seeds:
            model=set_model(model_name,params)
            model.fit(x_train,y_train)
            val_preds=model.predict_proba(x_val)[:,1]
            preds=model.predict_proba(x_test)[:,1]
            
            oof_preds[val_idx]+=val_preds/len(seeds)
            test_preds+=preds/len(seeds)/cv
        print(f"AUC is {roc_auc_score(y_val,oof_preds[val_idx]):.5f}\n")
    print(f"ðŸ”¥overall AUC is {roc_auc_score(train[TARGET],oof_preds):.5f}ðŸš€")
    return oof_preds,test_preds
```

```python
def to_submission(oof_preds,test_preds,desc):
    sub_["loan_paid_back"]=test_preds
    sub_.to_csv(f"./ensemble/{desc}_test.csv",index=False)

    df={"id":np.arange(len(oof_preds)),
       "oof_preds":oof_preds}
    df=pd.DataFrame(df)
    df.to_csv(f"./ensemble/{desc}_oof.csv",index=False)
```

```python
# lgb
oof_preds,test_preds=cv_train("lgb",lgb_params)
```

```python
os.makedirs("./ensemble",exist_ok=True)
to_submission(oof_preds,test_preds,"Single_lgb_baseline")
```
