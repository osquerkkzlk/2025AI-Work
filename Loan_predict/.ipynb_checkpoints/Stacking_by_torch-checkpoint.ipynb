# Stacking

```python
import torch
from torch import nn
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import StratifiedKFold
from torch.utils.data import DataLoader,TensorDataset
kf=StratifiedKFold(random_state=42,n_splits=5,shuffle=True)
```

```python
def construct_model(shape):
    model=nn.Sequential(
        nn.Linear(shape,32),
        nn.ReLU(inplace=True),
        nn.Dropout(0.1),

        nn.Linear(32,16),
        nn.ReLU(inplace=True),
        nn.Dropout(0.2),

        nn.Linear(16,1)
    )

    def init_weight(m):
        if isinstance(m,nn.Linear):
            nn.init.xavier_uniform_(m.weight)
            nn.init.zeros_(m.bias)
    return model.apply(init_weight)

class Accumulator():
    def __init__(self,count):
        self.metric=[0]*count

    def add(self,*args):
        for i,arg in enumerate(args):
            self.metric[i]+=arg
    def __getitem__(self,idx):
        return self.metric[idx]
        
class Recoder():
    def __init__(self,count):
        self.metric=[[]for _ in range(count)]

    def add(self,*args):
        for i,arg in enumerate(args):
            self.metric[i].append(arg)
    def __getitem__(self,idx):
        return self.metric[idx]
        
def predict(model,criterion,data_iter):
    model.eval()
    y_pred_probas,ys=[],[]
    metric=Accumulator(2)
    
    for x,y in data_iter:
        x,y=x.to(device),y.to(device)
        y_pred=model(x)
        
        y_pred_proba=torch.sigmoid(y_pred)
        y_pred_probas.append(y_pred_proba.cpu().numpy())
        ys.append(y.cpu().numpy())        
        # 
        metric.add(criterion(y_pred,y).cpu().item()*len(y),len(y))
    y_pred_probas=np.concatenate(y_pred_probas,axis=0)
    ys=np.concatenate(ys,axis=0)
    model.train()
    #    ()
    return y_pred_probas,metric[0]/(metric[1]+1e-8),ys
```

```python
# data
test_preds=np.zeros(len(oof_test))
labels=train.loan_paid_back
device="cuda" if torch.cuda.is_available() else "cpu"
epochs=10


# cvcv
for i,(train_idx,val_idx) in enumerate(kf.split(oof_train,labels)):
    print(f"{i+1}/5 is training!\n")
    data_train,data_val=torch.tensor(oof_train.iloc[train_idx].values,dtype=torch.float32),\
                            torch.tensor(oof_train.iloc[val_idx].values,dtype=torch.float32)
    target_train,target_val=torch.tensor(labels.iloc[train_idx].values,dtype=torch.float32).reshape(-1,1),\
                            torch.tensor(labels.iloc[val_idx].values,dtype=torch.float32).reshape(-1,1)
    data_test=torch.tensor(oof_test.values,dtype=torch.float32)
    # 
    train_iter=DataLoader(TensorDataset(data_train,target_train),batch_size=64,shuffle=True)
    val_iter=DataLoader(TensorDataset(data_val,target_val),batch_size=64,shuffle=False)
    # predict 
    test_iter=DataLoader(TensorDataset(data_test,torch.zeros(len(data_test),1)),batch_size=64,shuffle=False)
    
    model=construct_model(data_train.shape[1]).to(device)
    criterion=nn.BCEWithLogitsLoss()
    optim=torch.optim.AdamW(model.parameters(),lr=0.002,weight_decay=1e-4)
    scheduler=torch.optim.lr_scheduler.ReduceLROnPlateau(patience=3,factor=0.3,mode="max",optimizer=optim)
    
    metric_all=Recoder(2)
    
    for epoch in range(epochs):
        print(f"{epoch+1}/{epochs} is training !")
 
        metric=Accumulator(2)
        for x,y in train_iter:
            x,y=x.to(device),y.to(device)
            optim.zero_grad()
            pred=model(x)
            loss=criterion(pred,y)
            loss.backward()
            optim.step()
    
            metric.add(loss.cpu().item()*len(y),len(y))
        with torch.no_grad():
            y_pred_probas,temp_loss,ys=predict(model,criterion,val_iter)
            roc_score=roc_auc_score(ys,y_pred_probas)
            
        print(f"<train_loss> {metric[0]/metric[1]}    <val loss> {temp_loss}    <roc_auc_score> {roc_score}\n")
        scheduler.step(roc_score)
        metric_all.add(metric[0]/metric[1],temp_loss)
    
    print("ploting loss curve...")
    plt.title(f"{epoch+1}/{epochs}")
    plt.plot(np.arange(1,1+epochs),metric_all[0],"r-o",label="train loss")
    plt.plot(np.arange(1,1+epochs),metric_all[1],"b-o",label="val loss")
    plt.legend()
    plt.savefig("./Loss Curve.png")
    plt.show()


    print("predicting...")
    with torch.no_grad():
        test_preds+=predict(model,criterion,test_iter)[0]/5
```

```python
def to_submission(oof_preds,test_preds,desc):
    sub_["loan_paid_back"]=test_preds
    sub_.to_csv(f"./ensemble/{desc}_test.csv",index=False)

    df={"id":np.arange(len(oof_preds)),
       "oof_preds":oof_preds}
    df=pd.DataFrame(df)
    df.to_csv(f"./ensemble/{desc}_oof.csv",index=False)
```

```python
to_submission(np.zeros(0),test_preds,"Stacked_NN")
```
